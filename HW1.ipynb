# A. Calculate the gradient for the following functions:

# We can also implement gradient function in python

```python
from sympy import symbols , sympify
import numpy as np
def get_gradient():
    # getting inputs
    eq = input("enter equation: ")
    vars = input("enter variables with space between them: ")
    # handling inputs
    variables = symbols(vars)
    equation = sympify(eq, evaluate=True)
    gradient = []
    # getting derivative for each var
    for var in variables:
        gradient.append(equation.diff(var, 1))
    print(f"∇f = \n{np.array(gradient).reshape(len(gradient),1)}")
```

# B. Find the intervals of convexity and concavity of the function 𝑓(𝑥) = 2𝑥 3 − 18𝑥 2

# We can also find concave and convex interval using python

```python
import math
from sympy import sympify, symbols
from sympy.solvers import solve

def get_conv_conc_intervals():
    # getting inputs
    eq = input("Enter equation: ")
    # handling inputs
    equation = sympify(eq, evaluate=True)
    x = symbols('x')
    # initializing variables
    convex_interval = "The function has not convex interval"
    concave_interval = "The function has not concave interval"
    # getting second derivative
    second_diff = equation.diff(x, 2)
    cand_inf = solve(second_diff, x)[0]  # This is candidate inflation point
    # getting convex interval
        #substitute for x larger than candidate inflation point
    if second_diff.subs(x, cand_inf+1) > 1:
        convex_interval = (cand_inf, sympify(math.inf))
        #substitute for x smaller than candidate inflation point
    elif second_diff.subs(x, cand_inf-1) > 1:
        convex_interval = (sympify(-math.inf), cand_inf)
    # getting concave interval
        #substitute for x larger than candidate inflation point
    if -1 * second_diff.subs(x, cand_inf+1) > 1:
        concave_interval = (cand_inf, sympify(math.inf))
        #substitute for x smaller than candidate inflation point
    elif -1 * second_diff.subs(x, cand_inf-1) > 1:
        concave_interval = (sympify(-math.inf), cand_inf)
    print(f"Convex interval: {convex_interval}")
    print(f"Concave interval: {concave_interval}")
```

```python
get_conv_conc_intervals()
```

# C. Python code for check the convexity of the function and return the interval range if the function is convex

```python
def check_convexity():
    # handling inputs
    eq = input("Enter equation: ")
    equation = sympify(eq, evaluate=True)
    x = symbols('x')
    # initializing variables
    second_diff = equation.diff(x, 2)
    convex_interval = "The function has not convex interval"
    # check convexity of function
    if second_diff.subs(x, x) == second_diff.subs(x, 1): # True means it's a unit func
        status = "Not a convex function"
    else: # False means it's in terms of X
        status = "Convex Function"
        # calculate interval
        cand_inf = solve(second_diff, x)[0]      # This is candidate inflation point
        if second_diff.subs(x, cand_inf+1) > 1:
            convex_interval = (cand_inf, sympify(math.inf))
        elif second_diff.subs(x, cand_inf-1) > 1:
            convex_interval = (sympify(-math.inf), cand_inf)
    print(f"Status: {status}\nConvex Interval: {convex_interval}")
```

```python
check_convexity()
```
