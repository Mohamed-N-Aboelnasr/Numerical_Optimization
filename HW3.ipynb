# Batch gradient descent

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1400, 1600, 1700, 1875, 1100, 1550, 2350, 2450, 1425, 1700]) # x-------> House Size
x = x.reshape(len(x),1)
y = np.array([245000, 312000, 279000, 308000, 199000, 219000, 405000, 324000,319000,255000]) # y-----> House Price
y = y.reshape(len(x),1)
plt.scatter(x,y)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
theta_0 = 0
theta_1 = 0
alpha = 0.00000001
y_predict = theta_0 + theta_1 * x
m=len(x)
cost_func=(1/(2*m))*sum((y_predict-y)**2)
loss_list=[]
theta0_list=[]
theta1_list=[]
for i in range(200):
    y_predict = theta_0 + theta_1 * x
    theta0_grd = (1/m) * np.sum(y_predict - y)
    theta1_grd = (1/m) * np.sum((y_predict - y) * x)
    theta_0 = theta_0 - alpha * theta0_grd
    theta0_list.append(theta_0)
    theta_1 = theta_1 - alpha * theta1_grd
    theta1_list.append(theta_1)
    cost_func=(1/(2*m))*sum((y_predict-y)**2)
    loss_list.append(cost_func)
    
print(f"Theta_0 = {theta_0}, Theta_1 = {theta_1}")
print(f"Predicted Values \n {y_predict}")
```

```python
plt.plot(loss_list)
plt.xlabel("Iterations")
plt.ylabel("Loss Function")
plt.show()
```

```python
plt.plot(theta0_list,loss_list)
plt.xlabel("Theta 0")
plt.ylabel("Loss Function")
plt.show()
```

```python
plt.plot(theta1_list,loss_list)
plt.xlabel("Theta 1")
plt.ylabel("Loss Function")
plt.show()
```

```python
# Before fitting 
plt.scatter(x, y,label="Observed Data")
# After fitting
plt.plot(x, y_predict,label="Model", color='r')
plt.legend()
plt.show()
```

```python
# Evaluting the fitting of the model
y_predict = theta_0 + theta_1 * x
y_mean = np.mean(y)
SStot = np.sum((y - y_mean)**2)
SSres = np.sum((y - y_predict)**2)
R2 = 1 - (SSres / SStot)
print(f"R2: {R2}")
```

# Mini Batch gradient descent

```python
x = np.array([1400, 1600, 1700, 1875, 1100, 1550, 2350, 2450, 1425, 1700]) # x-------> House Size
x = x.reshape(len(x),1)
y = np.array([245000, 312000, 279000, 308000, 199000, 219000, 405000, 324000,319000,255000]) # y-----> House Price
y = y.reshape(len(x),1)
plt.scatter(x,y)
plt.xlabel("X")
plt.ylabel("Y predicted")
plt.show()
```

```python
theta_0 = 0
theta_1 = 0
alpha = 0.00000001
y_predict =[]
b=2  # we take batch size = 2 data points
u=int(len(x)/b) # no of updates = no of all data points / no of batch points
cost_func=0
loss_list=[]
theta0_list=[]
theta1_list=[]

for i in range(100): # this is number of epochs
    for j in range(u): # u = no of updates ---------> we update after each batch size
        y_predict = theta_0 + theta_1 * x[j*b:(j+1)*b]
        theta0_grd = (1/b) * np.sum(y_predict - y[j*b:(j+1)*b])
        theta1_grd = (1/b) * np.sum((y_predict - y[j*b:(j+1)*b]) * x[j*b:(j+1)*b])
        theta_0 = theta_0 - alpha * theta0_grd
        theta0_list.append(theta_0)
        theta_1 = theta_1 - alpha * theta1_grd
        theta1_list.append(theta_1)
        cost_func=(1/(2*b))*sum((y_predict-y[j*b:(j+1)*b])**2)
        loss_list.append(cost_func)
print(f"Cost Functions through iterations : \n {loss_list}")
```

```python
plt.plot(loss_list)
plt.xlabel("Iterations")
plt.ylabel("Loss Function")
plt.show()
```

```python
plt.plot(theta0_list,loss_list)
plt.xlabel("Theta 0")
plt.ylabel("Loss Function")
plt.show()
```

```python
plt.plot(theta1_list,loss_list)
plt.xlabel("Theta 1")
plt.ylabel("Loss Function")
plt.show()
```

```python
y_predict = theta_0 + theta_1 * x
# Before fitting 
plt.scatter(x, y,label="Observed Data")
# After fitting
plt.plot(x, y_predict,label="Model", color='r')
plt.legend()
plt.show()
```

```python
# Evaluting the fitting of the model
y_predict = theta_0 + theta_1 * x
y_mean = np.mean(y)
SStot = np.sum((y - y_mean)**2)
SSres = np.sum((y - y_predict)**2)
R2 = 1 - (SSres / SStot)
print(f"R2: {R2}")
```

# Stochastic gradient descent

```python
x = np.array([1400, 1600, 1700, 1875, 1100, 1550, 2350, 2450, 1425, 1700]) # x-------> House Size
x = x.reshape(len(x),1)
y = np.array([245000, 312000, 279000, 308000, 199000, 219000, 405000, 324000,319000,255000]) # y-----> House Price
y = y.reshape(len(x),1)
plt.scatter(x,y)
plt.xlabel("X")
plt.ylabel("Y predicted")
plt.show()
```

```python
theta_0 = 0
theta_1 = 0
alpha = 0.00000001
y_predict = 0
cost_func=0
loss_list=[]
theta0_list=[]
theta1_list=[]
for i in range(10): # this is the number of epochs
    for j in range(len(x)): # this is number of all data points and number of updates for each epoch
        y_predict = theta_0 + theta_1 * x[j]
        theta0_grd = 2*np.sum(y_predict - y[j])
        theta1_grd = 2*np.sum((y_predict - y[j]) * x[j])
        theta_0 = theta_0 - alpha * theta0_grd
        theta0_list.append(theta_0)
        theta_1 = theta_1 - alpha * theta1_grd
        theta1_list.append(theta_1)
        cost_func=sum((y_predict-y[j])**2)
        loss_list.append(cost_func)
    
print(loss_list)
```

```python
plt.plot(loss_list)
plt.xlabel("Iterations")
plt.ylabel("Loss Function")
plt.show()
```

```python
plt.plot(theta0_list,loss_list)
plt.xlabel("Theta 0")
plt.ylabel("Loss Function")
plt.show()
```

```python
plt.plot(theta1_list,loss_list)
plt.xlabel("Theta 1")
plt.ylabel("Loss Function")
plt.show()
```

```python
y_predict = theta_0 + theta_1 * x
# Before fitting 
plt.scatter(x, y,label="Observed Data")
# After fitting
plt.plot(x, y_predict,label="Model", color='r')
plt.legend()
plt.show()
```

```python
# Evaluting the fitting of the model
y_predict = theta_0 + theta_1 * x
y_mean = np.mean(y)
SStot = np.sum((y - y_mean)**2)
SSres = np.sum((y - y_predict)**2)
R2 = 1 - (SSres / SStot)
print(f"R2: {R2}")
```

# Batch gradient descent Based on momentum

```python
x = np.array([1400, 1600, 1700, 1875, 1100, 1550, 2350, 2450, 1425, 1700]) # x-------> House Size
x = x.reshape(len(x),1)
y = np.array([245000, 312000, 279000, 308000, 199000, 219000, 405000, 324000,319000,255000]) # y-----> House Price
y = y.reshape(len(x),1)
plt.scatter(x,y)
plt.xlabel("X")
plt.ylabel("Y predicted")
plt.show()
```

```python
theta_0 = 0
theta_1 = 0
alpha = 0.00000001
vt_0=0
vt_1=0
gamma=0.9
y_predict = theta_0 + theta_1 * x
m=len(x)
cost_func=(1/(2*m))*sum((y_predict-y)**2)
loss_list=[]
theta0_list=[]
theta1_list=[]
for i in range(100): # As we know we can decrease no of epochs to reach the same minimum loss such as in GD without Momentum
    y_predict = theta_0 + theta_1 * x
    theta0_grd = (1/m) * np.sum(y_predict - y)
    theta1_grd = (1/m) * np.sum((y_predict - y) * x)
    vt_0=gamma*vt_0+alpha*theta0_grd
    vt_1=gamma*vt_1+alpha*theta1_grd
    theta_0 = theta_0 - vt_0
    theta0_list.append(theta_0)
    theta_1 = theta_1 - vt_1
    theta1_list.append(theta_1)
    cost_func=(1/(2*m))*sum((y_predict-y)**2)
    loss_list.append(cost_func)
    
print(f"Theta_0 = {theta_0}, Theta_1 = {theta_1}")
print(f"Predicted Values \n {y_predict}")
```

```python
# As we see we can also decrease the no of epochs to just 10 and
#get almost same accuracy but we increase no of epochs to overcome 
#the oscillations around the minima and get excactly the same minima as in Batch GD 
plt.plot(loss_list)         
plt.xlabel("Iterations")
plt.ylabel("Loss Function")
plt.show()
```

```python
plt.plot(theta0_list,loss_list)
plt.xlabel("Theta 0")
plt.ylabel("Loss Function")
plt.show()
```

```python
plt.plot(theta1_list,loss_list)
plt.xlabel("Theta 1")
plt.ylabel("Loss Function")
plt.show()
```

```python
# Before fitting 
plt.scatter(x, y,label="Observed Data")
# After fitting
plt.plot(x, y_predict,label="Model", color='r')
plt.legend()
plt.show()
```

```python
# Evaluting the fitting of the model
y_predict = theta_0 + theta_1 * x
y_mean = np.mean(y)
SStot = np.sum((y - y_mean)**2)
SSres = np.sum((y - y_predict)**2)
R2 = 1 - (SSres / SStot)
print(f"R2: {R2}")
```

# Implementation of NAG for Stochastic gradient descent

```python
x = np.array([1400, 1600, 1700, 1875, 1100, 1550, 2350, 2450, 1425, 1700]) # x-------> House Size
x = x.reshape(len(x),1)
y = np.array([245000, 312000, 279000, 308000, 199000, 219000, 405000, 324000,319000,255000]) # y-----> House Price
y = y.reshape(len(x),1)
plt.scatter(x,y)
plt.xlabel("X")
plt.ylabel("Y predicted")
plt.show()
```

```python
theta_0 = 0
theta_1 = 0
alpha = 0.00000001
vt_0=0
vt_1=0
gamma=0.9
theta0_temp=0
theta1_temp=0
y_predict = 0
cost_func=0
loss_list=[]
theta0_list=[]
theta1_list=[]
for i in range(2): # this is the number of epochs
    for j in range(len(x)): # this is number of all data points and number of updates for each epoch
        # As we know we can decrease no of epochs to reach almost the same minimum loss such as in GD with Momentum
        theta0_temp=theta_0-gamma*vt_0 # here we moved to temp theta0 just using momentum
        theta1_temp=theta_1-gamma*vt_1 # here we moved to temp theta1 just using momentum
        y_predict = theta0_temp + theta1_temp * x[j] # here we predict y value for using it to get the gradient of temp thetas
        theta0_future_grd = 2*np.sum(y_predict - y[j]) # here we get the gradient of future theta0 which is based on just momentum
        theta1_future_grd = 2*np.sum((y_predict - y[j]) * x[j]) # here we get the gradient of future theta1 which is based on just momentum
        vt_0=gamma*vt_0+alpha*theta0_future_grd #here we get the mometum0 term based on the current momentum0 and gradient of future theta0
        vt_1=gamma*vt_1+alpha*theta1_future_grd ##here we get the mometum1 term based on the current momentum1 and gradient of future theta1
        theta_0 = theta_0 - vt_0 # here the actual next theta0
        theta0_list.append(theta_0) 
        theta_1 = theta_1 - vt_1 # here the actual next theta1
        theta1_list.append(theta_1)
        y_predict = theta_0 + theta_1 * x[j] # the actual prediction of y
        cost_func=sum((y_predict-y[j])**2)
        loss_list.append(cost_func)
    
print(loss_list)
```

```python
# here we can see the oscillations about minima is dramaticaly decreased
plt.plot(loss_list)
plt.xlabel("Iterations")
plt.ylabel("Loss Function")
plt.show()
```

```python
plt.plot(theta0_list,loss_list)
plt.xlabel("Theta 0")
plt.ylabel("Loss Function")
plt.show()
```

```python
plt.plot(theta1_list,loss_list)
plt.xlabel("Theta 1")
plt.ylabel("Loss Function")
plt.show()
```

```python
y_predict = theta_0 + theta_1 * x
# Before fitting 
plt.scatter(x, y,label="Observed Data")
# After fitting
plt.plot(x, y_predict,label="Model", color='r')
plt.legend()
plt.show()
```

```python
# Evaluting the fitting of the model
y_predict = theta_0 + theta_1 * x
y_mean = np.mean(y)
SStot = np.sum((y - y_mean)**2)
SSres = np.sum((y - y_predict)**2)
R2 = 1 - (SSres / SStot)
print(f"R2: {R2}")
```
