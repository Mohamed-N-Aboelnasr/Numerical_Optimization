## 2- using python code to plot the relationship between x and y before the fitting and after the fitting

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1400, 1600, 1700, 1875, 1100, 1550, 2350, 2450, 1425, 1700]) # x-------> House Size
x = x.reshape(len(x),1)
y = np.array([245000, 312000, 279000, 308000, 199000, 219000, 405000, 324000,319000,255000]) # y-----> House Price
y = y.reshape(len(x),1)
plt.scatter(x,y)
plt.show()
```

```python
theta_0 = 0
theta_1 = 500
alpha = 0.0000001
m = len(x)

for i in range(10000):
    y_pred = theta_0 + theta_1 * x
    theta0_grd = (1/m) * np.sum(y_pred - y)
    theta1_grd = (1/m) * np.sum((y_pred - y) * x)
    theta_0 = theta_0 - alpha * theta0_grd
    theta_1 = theta_1 - alpha * theta1_grd
    
print(f"Theta_0 = {theta_0}, Theta_1 = {theta_1}")
print(f"Predicted Values \n {theta_0 + theta_1 * x}")
```

```python
# Before fitting 
plt.scatter(x, y)
# After fitting
plt.plot(x, theta_0 + theta_1 * x, 'g')
plt.show()
```

```python
# Evaluting the fitting of the model
y_pred_final = theta_0 + theta_1 * x
y_mean = np.mean(y)
SStot = np.sum((y - y_mean)**2)
SSres = np.sum((y - y_pred_final)**2)
R2 = 1 - (SSres / SStot)
print(f"R2: {R2}")
```
